a. What is the difference between classification and clustering?
    
    Classification is a type of problem in which a set of predefined classes are known and the goal is to find the class that new object belongs to. In classification the labels are known.

    Clustering is a technique used to attmpt to group sets of objects and infer relationships based on patterns that an algorithm learns. In clustering the labels are not known. 


b. What is the difference between data warehouse and database?

    Data warehouses and databases are different in their functionalities. Data warehouses are used for data modeling and are optimized for reading operations. Databases are used for relational modeling and are optimized for writing operations.They also differ in that data warehouses have de-normalized data while databases have normalized data. Databases are used for OLTP and data warehouses are used for OLAP.

c. What is the difference between data mining and OLAP?

    Data mining and OLAP are very similar to each other. Perhaps their biggest difference is the way they operate on data. Data mining has a lower level of detail which allows for more detailed patterns to be seen in data. OLAP aggregates data on a high level of detail, allowing for more summary calculations.

```python
#Import packages
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import data
url = "https://github.com/PixarJunkie/fundamentals-of-data-science/raw/master/data/iris.csv" 
df = pd.read_csv(url)
df.head()
```

##### Problem 4-a

I think a classification algorithm using just these two variables would do decently. It would have a hard time classifying the difference between the Iris-versicolor and the Iris-virginica since there is a lot of overlap. From the scatterplot, it looks like a classification algorithm would classify the Iris-setosa quite well. 

```python
#sepal_length vs. sepal_width plot
sns.FacetGrid(df, hue = 'iris_class', size = 6).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
plt.show()
```

##### Problem 4-b

A classification algorithm using just these two variables would perform much better than the sepal variables. It would have the same problem with mis-classifying between the Iris-versicolor and the Iris-virginica in some ranges. Overall though, there are quite noticable differences between the three Iris classes and their respective pedal lengths/widths.

```python
#petal_length vs. petal_width plot
sns.FacetGrid(df, hue = 'iris_class', size = 6).map(plt.scatter, "petal_length", "petal_width").add_legend()
sns.plt.show()
```

##### Problem 4-c

The sepal length and sepal width are the most "normally" distributed. The petal length appears to by bimodal while the petal width seems to be multimodal.

```python
for col in df.iloc[:, :4]:
    df.hist(col)
    plt.show()
```

##### Problem 4-d

From looking at all sepal lengths falling outside the 2.7th standard deviation, there were no sepal lengths outside that range. This would indicate that there are no statistical outliers for the sepal length variable.

```python
#Creating boolean dataframe of sepal lengths outside the 2.7 standard deviation
outlier_std = 2.7

outlier_trans = df[['sepal_length', 'iris_class']].groupby('iris_class').transform(lambda x: (x - x.mean()).div(x.std()))

outlier_df = outlier_trans.abs() > outlier_std
outlier_df[outlier_df.sepal_length == True]
```

##### Problem 4-e

Using the same method as above, the conclusion for petal length was the same. Since there were not petal lengths outside the 2.7th standard deviation, there are no statistical outliers.

```python
#Creating boolean dataframe of petal lengths outside the 2.7 standard deviation
outlier_std = 2.7

outlier_trans = df[['petal_length', 'iris_class']].groupby('iris_class').transform(lambda x: (x - x.mean()).div(x.std()))

outlier_df = outlier_trans.abs() > outlier_std
outlier_df[outlier_df.petal_length == True]
```
